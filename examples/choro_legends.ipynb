```python
import geopandas
from geopandas.datasets import get_path
from geopandas import read_file
```

```python
pth = get_path('naturalearth_lowres')
```

```python
df = read_file(pth)
```

```python
%matplotlib inline
ax = df.plot(column='pop_est', scheme='QUANTILES', k=3, \
             cmap='OrRd', legend=True)
```

```python
labels = [t.get_text() for t in ax.get_legend().get_texts()]
```

```python
labels
```

```python
ax = df.plot(column='pop_est', scheme='percentiles', k=3,
                          classification_kwds={'pct': [50, 100]}, cmap='OrRd',
                          legend=True)
labels = [t.get_text() for t in ax.get_legend().get_texts()]
labels
```

```python
import numpy as np
df['NEGATIVES'] = np.linspace(-10, 10, len(df.index))

ax = df.plot(column='NEGATIVES', scheme='FISHER_JENKS', k=3,
                          cmap='OrRd', legend=True)
labels = [t.get_text() for t in ax.get_legend().get_texts()]
labels
```

```python
ax = df.plot(column='pop_est', scheme='percentiles', k=3,
                          classification_kwds={'pct': [50, 100]}, cmap='OrRd',
                          legend=True, fmt="{:.1f}")
labels = [t.get_text() for t in ax.get_legend().get_texts()]
labels
```

```python
ax = df.plot(column='pop_est', scheme='percentiles', k=3,
                          classification_kwds={'pct': [50, 100]}, cmap='OrRd',
                          legend=True, fmt="{0:.2e}")
labels = [t.get_text() for t in ax.get_legend().get_texts()]
labels
```

```python
ax = df.plot(column='pop_est', scheme='percentiles', k=3,
                          classification_kwds={'pct': [50, 100]}, cmap='OrRd',
                          legend=True, fmt="{0:.2e}", legend_kwds={'loc': 'lower right'})
labels = [t.get_text() for t in ax.get_legend().get_texts()]
labels
```
